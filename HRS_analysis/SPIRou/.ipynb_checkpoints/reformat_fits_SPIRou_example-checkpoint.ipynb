# You need to separate out the files from the extended headers from the SPIROU archive in order to run STARSHIPS

e files and t files are the most important here

e files: 2D extracted spectra that use the instrument profile and order localization and performs optimal extraction

t files: 2D spectra in the same format than e spectra, after the correction of telluric lines has been applied.

# Adding info to the primary header
Starships wants to read certain things from the apero pipeline in the primary header that are not there but are in a different extension. So we can add them to our files

```python
import glob
#run this for all the files that need the informaton added

list_of_filenames = sorted(glob.glob('/home/mathisb/Github/HRS_data/SPIRou/TRAPPIST-1/*n_AB.fits'))

for i in list_of_filenames:
    # with fits.open(i) as hdu:
    #     print(hdu[1].header['BJD'])

    with fits.open(i, mode='update') as filehandle:
        filehandle[1].header['CDBBLAZE']= filehandle[0].header['CDBBLAZE']
        filehandle[1].header['WAVEFILE']= filehandle[0].header['WAVEFILE']
        filehandle[1].header['BJD']= filehandle[0].header['BJD']
        filehandle[1].header['BERV']= filehandle[0].header['BERV']
```

```python
#you can double check here that the primary header now has the added infomation fom the first extension

for i in list_of_filenames:
    with fits.open(i) as hdu:
        print(hdu[0].header['obsid'], "\t", hdu[1].header['WAVEFILE'], hdu[0].header['WAVEFILE'])
```

# Lets Create a "Recon file"....In this case the RECON FILE IS the tellurics

This needs to be created for every single file, since it will be different in each file. we can prove this below. 

```python
# Plot tellurics
list_of_filenames = sorted(glob.glob('/home/mathisb/Github/HRS_data/SPIRou/TRAPPIST-1/*t.fits'))


for i in list_of_filenames:
    # print(orig.info())
    with fits.open(i) as orig:
        
        wave=orig[2].data
        tell=orig[4].data
        
        plt.plot(wave[42],tell[42])
 
 #clearly it is changing thoughout the observation as airmass changes the telluric lines
     
```

```python
for i in list_of_filenames:
    # print(orig.info())
    with fits.open(i) as orig:
        

        primary_header = (orig[0].header)
        obsid=orig[0].header['obsid']
        obsid_str = str(obsid)
        
        
        empty_primary = fits.PrimaryHDU(header=primary_header)
        # image_hdu = fits.ImageHDU(orig[5].data)
    
        hdul = fits.HDUList([empty_primary])
        
        out= '/home/mathisb/Github/HRS_data/SPIRou/TRAPPIST-1/'+ obsid_str +'_RECON.fits'
        # print(out)
        hdul.writeto(out)
        
        fits.append(out, orig[4].data, orig[4].header)
```

## Time of mid-transit

```python
import numpy as np
from astropy.time import Time, TimeDelta
import astropy.units as u
from datetime import datetime

t = Time(2458362.0021, format="jd", scale='tdb')  # mid-transit time from Hirano et al. 2020

for n in [763]:
    print(t + n * 1.51087 * u.day)
```

```python
# HIRANO: 2459514.7959100003
# AGOL: 2459514.805984833954 (taking into consideration TTV)

# Difference:
print(2459514.805984833954 - 2459514.7959100003)
```

```python
print(TimeDelta(2459514.805984833954, -2459514.7959100003, format="jd", scale='tdb').to_value(format="sec") / 60 * u.min)
```
