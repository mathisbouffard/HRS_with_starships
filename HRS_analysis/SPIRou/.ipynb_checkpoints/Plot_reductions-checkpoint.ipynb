# Get planet info and determine directories to use

```python
#--------------------------------------------------------------------------------------------
# Planet, reduction and instrument
pl_name = 'TRAPPIST-1 b'
reduction = 'v006'
instrument = "SPIRou"

#--------------------------------------------------------------------------------------------

# Directory where to save reductions
# > use scratch if available, use home if not
try:
    out_dir = Path(os.environ['SCRATCH'])
except KeyError:
    out_dir = Path.home()

# Output reductions in dedicated directory
pl_name_fname = ''.join(pl_name.split())
out_dir /= Path(f'HRS_reductions/{instrument}/{pl_name_fname}/{reduction}')
# Make sure the  directory exists
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
# Relevant directories

# Where to find the observations (data)?
obs_dir = "/home/mathisb/Github/HRS_data/SPIRou/TRAPPIST-1/"  # for TRAPPIST-1 SPIRou data

# Where to save figures?
path_fig = Path(f'HRS_reductions/SPIRou/{pl_name_fname}/Figures')
```

```python
# All the observations must be listed in files.
# > there are files in the observations directory in which the name of all observation files is written
# We need the e2ds, the telluric corrected and the reconstructed spectra.

list_filenames = {'list_e2ds': 'list_e2ds_visit_1.txt',
                  'list_tcorr': 'list_tcorr_visit_1.txt',
                  'list_recon': 'list_recon_visit_1.txt'}
```

```python
# SPECIFY PARAMETERS OF THE PLANET
# If not specified, the default parameter from exofile are taken. Make sure they are ok.

# Parameters from trappist.one website
ap     = 0.01150*u.au       # semi-major axis of planet
R_star = 0.121*u.R_sun      # radius of star
R_pl   = 1.127*u.R_earth    # radius of planet
M_star = 0.089*const.M_sun  # mass of star
e      = 0                  # eccentricity
w      = 4.712389*u.rad     # argument of periapsis (rad)
mid_tr = 2459514.805984833954*u.d  # mid-transit time (calculated from TTV BJD, Agol et al. 2021)



#--------------------------------------------------------------------------------------------


obs = Observations(name=pl_name, 
                   pl_kwargs = {'M_star': M_star, 
                                'R_star': R_star,
                                'ap': ap,
                                'R_pl': R_pl,
                                'mid_tr': mid_tr,
                                'excent': e,
                                'w': w})

# Get the data
obs.fetch_data(obs_dir, **list_filenames)
p = obs.planet
```
