# Read spectra and useful quantities

## Note: Better to use the scratch space to save the reductions.
You have infinite space, but the files are deleted if untouched for 2 months. It allows to save as many reductions as desired. Once the correct reduction parameters are set, you can either move them into your home directory.

```python
pl_name = 'WASP-33 b'
reduction = 'v07232'

# Use scratch if available, use home if not.
try:
    out_dir = Path(os.environ['SCRATCH'])
except KeyError:
    out_dir = Path.home()

# Output reductions in dedicated directory
pl_name_fname = ''.join(pl_name.split())
out_dir /= Path(f'HRS_reductions/SPIRou/{pl_name_fname}/{reduction}')
# Make sure the  directory exists
out_dir.mkdir(parents=True, exist_ok=True)

# --- Where to find the observations?
# The goal is to keep raw observation in a shared directory, with this struscture:
# shared_dir = Path.home() / Path('mathisb/projects/def-dlafre/shared')
# obs_dir = shared_dir / Path(f'SPIRou/{pl_name_fname}_{reduction}/')

# But you can also put them in your project account (like I did for WASP-33b)
proj_dir = Path.home() / Path('projects/def-ncowan')
obs_dir = proj_dir / Path(f'SPIRou_Observations/WASP-33b/{reduction}/{pl_name_fname}/')

# Where to save figures?
path_fig = Path(f'HRS_reductions/SPIRou/{pl_name_fname}/Figures')
```

```python
# All the observations must be listed in files.
# > there are files in the observations directory in which the name of all observation files is written
# We need the e2ds, the telluric corrected and the reconstructed spectra.

list_filenames = {'list_e2ds': 'list_e2ds_visit_2019-10-07',
                  'list_tcorr': 'list_tcorr_visit_2019-10-07',
                  'list_recon': 'list_recon_visit_2019-10-07'}
```

```python
# Some parameter of the planet system can be specified here.
# If not, the default parameter from exofile are taken. Make sure they are ok.
ap = 0.0259*u.au
R_star = (ap/3.69).to(u.R_sun)
R_pl = (0.1143*R_star).to(u.R_jup)
M_star = 1.561*const.M_sun
obs = Observations(name=pl_name, 
                   pl_kwargs = {'M_star':M_star, 
                                'R_star':R_star,
                                'ap' : ap,
                               'R_pl' : R_pl})

# Get the data
obs.fetch_data(obs_dir, **list_filenames)
```

```python
p = obs.planet
```

# Build transmission spectrum

```python
pl_obs.generate_all_transits?
```

```python
"""
param_all: Reduction parameters
telluric fraction to mask (usually varied between 0.2 and 0.5), 
limits for the wings (fixed at 0.97), 
width of the smoothing kernel for the low pass filter (fixed at 51), 
useless param, 
width of the gaussian kernel for low pass filter (fixed at 5),
nPC (nb of principal components) to remove (varied between 1 and 5),
sigma clips params (fixed at 5.0)

(So I basically only change tellu frac and nPC)
"""
visit_name = 'night1'
mask_wings = 0.9  # Mask wings of deep tellurics

n_RV_inj=151
corrRV0 = np.linspace(-150, 150, n_RV_inj)

coeffs = [0.532]
cbp=True  # Correct bad pixels
ld_model = 'linear'
kind_trans = 'emission'

RVsys = [-23.0]
iout_all = ['all']
polynome = [False] 
do_tr=[1]
transit_tags = [None]
kwargs_gen_tr = {
'coeffs' : coeffs,
'ld_model' : ld_model,
'do_tr' : do_tr,
'kind_trans' : kind_trans,
'polynome' : polynome,
'cbp': cbp
}

kwargs_build_ts = {
'clip_ratio' : 6,
'clip_ts' : 6,
'unberv_it' : True,
}

for n_pc in [1,3,5,10]:
    params_all=[[0.20, mask_wings, 51, 41, 5, n_pc, 5.0, 5.0, 5.0, 5.0]]
    list_tr = pl_obs.generate_all_transits(obs, transit_tags, RVsys, params_all, iout_all,
                                           **kwargs_gen_tr, **kwargs_build_ts)

    # Save sequence with all reduction steps
    out_filename = f'sequence_{n_pc}-pc_mask_wings{mask_wings*100:n}'
    pl_obs.save_single_sequences(out_filename, list_tr['1'], path=out_dir,
                                 filename_end=visit_name, save_all=True)
    
    # Save sequence with only the info needed for a retrieval (to compute log likelihood).
    out_filename = f'retrieval_input_{n_pc}-pc_mask_wings{mask_wings*100:n}'
    pl_obs.save_sequences(out_filename, list_tr, do_tr, path=out_dir)
```
